```python
humans = dataframe_plm[dataframe_plm["species"] == "Homo sapiens"]
#df2['id'].isin(['SP.POP.TOTL','NY.GNP.PCAP.CD'])
humans
```

```python
humans[humans["code"] == "Y"]
```

```python
target_seq_len = 33
target_left_right = 16
```

```python
def write_fasta(dataframe, file_path):
    dataframe = dataframe.drop_duplicates(subset=['sequence'])
    print("After duplicates removal")
    print(dataframe)
    data_fasta = ""
    for i, row in dataframe.iterrows():
        data_fasta += ">" + row["acc"] + "\n"
        data_fasta += row["sequence"] + "\n"
    
    with open(file_path, 'w') as f:
        f.write(data_fasta)
    return dataframe

r_Y_id = list()
r_Y_seq = list()

for i, row in humans.iterrows():
    pos = int(row["position"])
    if pos > 33 and pos + target_left_right < len(row["sequence"]):
        extracted_seq = row["sequence"][pos - target_left_right - 1: pos + target_left_right]
        if row["code"] == "Y":
            r_Y_id.append(row["acc"])
            r_Y_seq.append(extracted_seq)
        
            
Y_dataframe = pd.DataFrame(zip(r_Y_id, r_Y_seq), columns=["acc", "sequence"])

print(Y_dataframe)
```

```python
dedup_Y_dataframe = write_fasta(Y_dataframe, "../dataset/phospho.elm_dataset/phosphoELM_all_latest.dump/Phospho_pos_Y.fasta")  
dedup_Y_dataframe.to_csv("../dataset/phospho.elm_dataset/phosphoELM_all_latest.dump/Phospho_pos_Y.tsv", sep="\t", index=None)
```

```python
# create negative dataset
Y_neg_id = list()
Y_neg_seq = list()

def extract_neg_seq(original_seq, s_id, pos, s_len=33):
    r_id = list()
    r_seq = list()
    for p in pos:
        if p > s_len:
            extracted_seq = original_seq[p - target_left_right: p + target_left_right + 1]
            if len(extracted_seq) >= s_len:
                r_id.append(s_id)
                r_seq.append(extracted_seq)
    return r_id, r_seq
                        

for i, row in humans.iterrows():
    if len(row["sequence"]) > 33:

        Y_pos = [pos for pos, char in enumerate(row["sequence"]) if char == "Y"]
        
        if row["code"] == "Y":
            # exclude POS position
            Y_pos = [ item for item in Y_pos if item != row["position"] - 1 ]
            ids, seqs = extract_neg_seq(row["sequence"], row["acc"], Y_pos)
            Y_neg_id.extend(ids)
            Y_neg_seq.extend(seqs)
        

Y_neg_dataframe = pd.DataFrame(zip(Y_neg_id, Y_neg_seq), columns=["acc", "sequence"])
```
