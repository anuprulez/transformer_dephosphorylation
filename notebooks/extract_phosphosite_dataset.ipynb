```python
import pandas as pd
import numpy as np
from Bio import SeqIO
phosphosite_df = pd.read_csv("../dataset/phosphosite_org/Phosphorylation_site_dataset.csv", sep="\t")
phosphosite_df
```

```python
posphosite_humans = phosphosite_df[phosphosite_df["ORGANISM"] == "human"]
posphosite_humans
```

```python
fasta_file = "../dataset/phosphosite_org/Phosphosite_PTM_seq.fasta"
p_site_s = "s"
p_site_t = "t"
target_seq_len = 33
target_left_right = 16

r_S_id = list()
r_S_seq = list()
r_T_id = list()
r_T_seq = list()

for seq_record in SeqIO.parse(fasta_file, "fasta"):
        data = seq_record
        record_split = seq_record.description.split("|")
        if record_split[2] == "human":
            seq = seq_record.seq
            for index, s in enumerate(seq):
                if s == p_site_s:
                    if index > 33 and index + target_left_right < len(seq):
                        extracted_seq = seq[index - target_left_right: index + target_left_right + 1]
                        extracted_seq = extracted_seq.upper()
                        r_S_id.append(record_split[3])
                        r_S_seq.append(str(extracted_seq))
                if s == p_site_t:
                    if index > 33 and index + target_left_right < len(seq):
                        extracted_seq = seq[index - target_left_right: index + target_left_right + 1]
                        extracted_seq = extracted_seq.upper()
                        r_T_id.append(record_split[3])
                        r_T_seq.append(str(extracted_seq))
        

S_dataframe = pd.DataFrame(zip(r_S_id, r_S_seq), columns=["acc", "sequence"])

T_dataframe = pd.DataFrame(zip(r_T_id, r_T_seq), columns=["acc", "sequence"])

print(S_dataframe)
print()
print(T_dataframe)
```

```python
S_dataframe = S_dataframe.drop_duplicates(subset=['sequence'])

T_dataframe = T_dataframe.drop_duplicates(subset=['sequence'])

print(S_dataframe)
print()
print(T_dataframe)
```

```python
def write_fasta(dataframe, file_path):
    dataframe = dataframe.drop_duplicates(subset=['sequence'])
    print("After duplicates removal")
    print(dataframe)
    data_fasta = ""
    for i, row in dataframe.iterrows():
        data_fasta += ">" + row["acc"] + "\n"
        data_fasta += row["sequence"] + "\n"
    
    with open(file_path, 'w') as f:
        f.write(data_fasta)
    return dataframe
```

```python
dedup_S_dataframe = write_fasta(S_dataframe, "../dataset/phosphosite_org/all_pos_S.fasta")  
dedup_S_dataframe.to_csv("../dataset/phosphosite_org/all_pos_S.tsv", sep="\t", index=None)
```

```python
dedup_T_dataframe = write_fasta(T_dataframe, "../dataset/phosphosite_org/all_pos_T.fasta")  
dedup_T_dataframe.to_csv("../dataset/phosphosite_org/all_pos_T.tsv", sep="\t", index=None)
```

```python
# create negative dataset
S_neg_id = list()
S_neg_seq = list()

T_neg_id = list()
T_neg_seq = list()

def extract_neg_seq(original_seq, s_id, pos, s_len=33):
    r_id = list()
    r_seq = list()
    for p in pos:
        if p > s_len:
            extracted_seq = original_seq[p - target_left_right: p + target_left_right + 1]
            if len(extracted_seq) >= s_len:
                r_id.append(s_id)
                r_seq.append(extracted_seq)
    return r_id, r_seq
                        

for seq_record in SeqIO.parse(fasta_file, "fasta"):
        data = seq_record
        record_split = seq_record.description.split("|")
        if record_split[2] == "human":
            seq = seq_record.seq
            for index, s in enumerate(seq):
                if s == "S" and s != "s":
                    if index > 33 and index + target_left_right < len(seq):
                        extracted_seq = seq[index - target_left_right: index + target_left_right + 1]
                        extracted_seq = extracted_seq.upper()
                        S_neg_id.append(record_split[3])
                        S_neg_seq.append(str(extracted_seq))
                if s == "T" and s != "t":
                    if index > 33 and index + target_left_right < len(seq):
                        extracted_seq = seq[index - target_left_right: index + target_left_right + 1]
                        extracted_seq = extracted_seq.upper()
                        T_neg_id.append(record_split[3])
                        T_neg_seq.append(str(extracted_seq))

S_neg_dataframe = pd.DataFrame(zip(S_neg_id, S_neg_seq), columns=["acc", "sequence"])
T_neg_dataframe = pd.DataFrame(zip(T_neg_id, T_neg_seq), columns=["acc", "sequence"])
```

```python
def write_tsv_fasta(file_path, save_to_file, max_rows=20000):
    fasta_text = ""
    dataframe = pd.read_csv(file_path, sep="\t")
    print(dataframe)
    for i, row in dataframe.iterrows():
        fasta_text += ">" + row[0] + "\n"
        fasta_text += row[1] + "\n"
        if i > max_rows:
            break
    with open(save_to_file, "w") as f:
        f.write(fasta_text)
```
